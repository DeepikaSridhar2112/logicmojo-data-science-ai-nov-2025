#### Question-7

Write a function `rotate_list(lst, k)` that rotates a given list lst to the right by `k` positions in-place. Assume `k` is a non-negative integer and can be greater than the length of the list.

Example:

```python
my_list = [1, 2, 3, 4, 5]
rotate_list(my_list, 2)
print(my_list)
# Expected output: [4, 5, 1, 2, 3]
```

````python
# Write your code here
def rotate_list(actual_list,number):
    rotated_list=[]
    if number>len(actual_list):
        number = number%len(actual_list)
    for i in range(len(actual_list)-1,len(actual_list)-number-1,-1):
        rotated_list.insert(0,actual_list[i])
    for i in range(0,len(actual_list)-number):
        rotated_list.append(actual_list[i])
    actual_list[:] = rotated_list
    return actual_list

my_list = [1, 2, 3, 4, 5]
rotate_list(my_list, 2)
print(my_list)    

````

Example:

```python
n = 5
# Expected output:
#     1
#    1 1
#   1 2 1
#  1 3 3 1
# 1 4 6 4 1
```

````python
# Write your code here
def pascal_triangle(depth):
    sequence = []
    for i in range(0,depth):
        numbers = [1]
        if(i>1):
            previous_list = sequence[i-1]
            for j in range(1,i):
                value = previous_list[j] + previous_list[j-1]
                numbers.append(value)
        if(i>0):
            numbers.append(1)
        sequence.append( numbers)
    return sequence
def format_response(input,depth):
    initial_space_count = depth
    for i in range (0,depth):
        values = input[i]
        print(" " * initial_space_count + " ".join(str(v) for v in values))
        initial_space_count-=1 
depth = int(input("Enter the depth"))    
result = pascal_triangle(depth)
format_output = format_response(result,depth)
        

````

#### Question-10

Write a function `reverse_the_words(s)` which will take a string `s` as input and reverses the words in the string. 

Example:

```python
s = "Python is awesome and challenging"
r = reverse_the_words(s)
print(r)
# Expected output: challenging and awesome is Python
```

````python
# Write your code here
def reverse_the_words(input):
    output = ""
    words= input.split()
    for i in range(len(words)-1,-1,-1):
      output+= words[i]
      if i!=0:
          output+= " "
    return output
s = "Python is awesome and challenging"
r = reverse_the_words(s)
print(r)
````

#### Bonus question

Write a function `word_counter(s)` which will take a string `s` as input and returns a dictionary with count of words (case insensitive). Remove all the punctuations and special symbols present in the string.  

Example:
```python
s = "This city is a beautiful city. This is where I live"
d = word_counter(s)

# Expected output (output can be in any order): {"this": 2, "city": 2, "is": 2, "a": 1, "beautiful": 1, "where": 1, "i": 1, "live": 1}
```

````python
# Write your code here
def word_counter(input_line):
    response={}
    words= input_line.split()
    for word in words:
     if word not in response:
        response[word]=1
     else:
        response[word] = response[word]+1
    return response
s = "This city is a beautiful city. This is where I live"
d = word_counter(s)
print(d)
````
