## Problem: 4

## Problem: 5

Create an 8 x 8 array with a checkerboard pattern of zeros and ones.

```python
# Write your code here
pattern = np.arange(8) % 2
second_pattern = np.arange(1,9,1)%2
arr= np.zeros((8,8),dtype=int)
for i in range(0,8):
    if(i%2 == 0):
        arr[i,:]=pattern
    else:
        arr[i,:]=second_pattern
print(arr)

```

## Problem: 7 
Create a function which creates an $n×n$ array with $(i , j)$ -entry equal to $i+ j$

```python
# Write your code here
def array_creator(size):
    # custom_array = np.zeros((size,size),dtype=int)
    # for i in range(0,size):
    #     for j in range(0,size):
    #         custom_array[i][j]=i+j
    # return custom_array
    rows = np.arange(size).reshape(size, 1)   # shape (5, 1)
    cols = np.arange(size).reshape(1, size)   # shape (1, 5)
    arr = rows + cols  
    return arr

arr = array_creator(6)
print(arr)

```
